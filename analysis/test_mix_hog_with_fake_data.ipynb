## Represent an object map using a simple mixture-of-multinomials model

#### create environment

```python
#create environment
import numpy as np
from PIL import Image
from object_parsing.src.image_objects import view_mask
from imagery_psychophysics.src.mixture_of_histograms import mix_hog_image, plain_plot
%pylab inline
```

#### model and simulation parameters

```python
##number of documents / probes
M = 500

##number of pixels per document
P = 50

##controls the uniformity of the initial guess at the prior distribution
UNIFORMITY = 1

##controls how "smooth" the model parameter estimates are
SMOOTHING = .0001  ## << usually want number less than 1, it seems
```

#### grab an image / object map

```python
cur_image = 3600
image_path = '/musc.repo/Data/shared/my_labeled_images/pictures/%0.6d.png'
mask_path = '/musc.repo/Data/shared/my_labeled_images/labeled_image_maps/%0.6d.png'
img = Image.open(image_path %(cur_image))
msk = Image.open(mask_path %(cur_image))
msk=msk.resize(size=(64,64))
figure()
plain_plot(gca())
title('original image')
imshow(img)
figure()
plain_plot(gca())
title('object mask downsampled to 32x32')
imshow(msk)
#view_mask(msk)
mask_array = np.array(msk)
N = mask_array.shape
V = np.prod(N)
print 'vocabulary size: %d' %(V)
```

#### represent the image mask as the parameters of a mixture-of-histograms generative model

```python
#K = number of objects
K = len(np.unique(mask_array))

##to create prior, count the number of pixels in each document
L = [sum(sum(mask_array==ii)).astype('double') for ii in np.arange(K)+1]
print 'total number of pixels in each document:'
print L
print

prior_dist = np.array(L/V).reshape((1,K)) ##fraction of pixels in each object
print 'object prior:' 
print prior_dist

##to create the likelihood distributions, treat each object mask as the parameters of a multinomial

##this just reformats the mask the way we want it to be
def format_mask_as_lkhd():
    return np.hstack([np.atleast_2d(np.array(mask_array==ii).flatten()).T for ii in np.arange(K)+1])

##divide by total number of pixels to get a true probability instead of a count
lkhd_dist = format_mask_as_lkhd()/L

##If any pixel has 0 likelihood, set it to a very tiny number. this avoids lots of irritating bugs.
lkhd_dist = lkhd_dist + np.finfo(np.double).tiny

##re-normalize to make sure each column still sums to 1
lkhd_dist = lkhd_dist/lkhd_dist.sum(axis=0, keepdims=True)


```

#### sample multiple "probes" from the "true " model

```python


probes = np.zeros((V,M))

##use prior to determine number of probes per object
probes_per_object = np.random.multinomial(M, np.squeeze(prior_dist))

##generate object patches
cnt = 0
for ii,nn in enumerate(probes_per_object):
    for jj in range(nn):
        dx = np.random.choice(np.arange(V), size=P, replace=False, p=np.atleast_1d(lkhd_dist[:,ii]))
        probes[dx, cnt] = 1
        cnt += 1

##check a few probes
figure()
plain_plot(gca())
title('first probe')
imshow(probes[:,0].reshape(N))
figure()
plain_plot(gca())
title('last probe')
imshow(probes[:,M-1].reshape(N))
```

#### initialize a naive model 

```python
##prior and likelihood distributions are randomly sample from a dirichlet distribution
naive_prior_dist = np.random.dirichlet(UNIFORMITY*np.ones(K)).reshape((1,K))
naive_lkhd_dist = np.random.dirichlet(np.ones(V), size=K).T

##instantiate a mixture of histograms model (image sub-class)
mhi = mix_hog_image(naive_prior_dist, naive_lkhd_dist)
print mhi.prior_dist
```

#### ...and then restimate parameters

```python
##EM
print('the log-likelihood of the data, given the current best estimate of the model parameters')
mhi.smoothing_param = SMOOTHING
mhi.run_em(probes)

```

#### visualize the model parameters and the MAP estimate of the object that each pixel belongs to

```python
print '(top) the likelihood distributions'
mhi.view_lkhd_dist()
print '(middle) the prior distribution'
mhi.view_prior()
print '(bottom) the best estimate of object membership for each pixel'
mhi.view_MAP()
```
