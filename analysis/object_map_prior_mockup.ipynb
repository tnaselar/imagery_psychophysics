#### the leaf indicator function

```python
def leaf_indicator(base_object, leaf):
  return np.array(base_object==leaf).all()
```

#### calculate inclusion probabilities for different levels of resolution

```python
inclusion_probability = [0.0]*len(resolutions)
cnt = 0
for stride in resolutions:
  for ii in range(len(leaf_list)):
    inclusion_probability[cnt] += leaf_indicator(base_object[::stride, ::stride], leaf_list[ii][::stride, ::stride])
  cnt += 1

inclusion_probability = np.array(inclusion_probability)
inclusion_probability /= len(off_x)
plot(resolutions, inclusion_probability)
xlabel('stride (inverse of resolution)')
ylabel('probability')
```

##### As resolution decreases it becomes easier to match the leafs to the base object

#### calculate probability of non-inclusion: this means evaluating the null base object

```python
null_base = np.zeros((500,500))
non_inclusion_probability = [0.0]*len(resolutions)
cnt = 0
for stride in resolutions:
  for ii in range(len(leaf_list)):
    non_inclusion_probability[cnt] += leaf_indicator(null_base[::stride, ::stride], leaf_list[ii][::stride, ::stride])
  cnt += 1

non_inclusion_probability = np.array(non_inclusion_probability)
non_inclusion_probability /= len(off_x)
plot(resolutions, non_inclusion_probability)
```

##### As resolution decreases, it becomes more likely that some of the leaves will fit in between the sample points.

#### Evaluation of full object map probabilities

$$P(Z) = \frac{1}{1-Q_{\sigma(Z,0)}}\sum_{n=1}^K Q_{\sigma(Z,n)}P(Z \backslash n) $$

#### A bunch of functions we'll need to evaluate P(Z)

```python
## a bunch of functions we'll need
from operator import mul    # or mul=lambda x,y:x*y
from fractions import Fraction

def nCk(n,k): 
  return int( reduce(mul, (Fraction(n-i, i+1) for i in range(k)), 1) )



def format_object(z):
    return z.reshape((z.size,1))

def number_of_objects(Z):
    return Z.shape[1]

#def inclusion_prob(base_object, leaves):
#    inclusion_probability = float(sum([1 for leaf in leaves if leaf_indicator(base_object, leaf)]))/len(leaves)
#    return inclusion_probability

def inclusion_prob(base_object, leaves):
    n_leaves = leaves.shape[1]
    inclusion_probability = np.sum(np.array(base_object==leaves).all(axis=0)).astype('float')/n_leaves
    return inclusion_probability


##note there is no 0th object
def pull(Z, n):
    if n:
        return np.atleast_2d(Z[:,n-1]).T
    else:
        return np.zeros((Z.shape[0],1))

def remove(leaves, Z, n):
    if n:
        n = n-1
        K = number_of_objects(Z)
        #indices = [ii for ii in range(K) if not ii == n] ##set difference 
        foo = leaves[~Z[:,n].astype('bool'), :]
        indices = foo.any(axis=0)
        return foo[:,indices]
    else:
        return Z

#def downsample_leaves(leaves, Z, n):
    ##downsample leaves. remove leaves that have been downsampled out of existence.
#    if n:
#        n = n-1
#        return filter(sum, [leaves[ii][~Z[:,n].astype('bool'),:] for ii in range(len(leaves))])
#    else:
#        return filter(sum, leaves)

def downsample_leaves(leaves, Z, n):
    return 

##the recursive prior
def object_map_prior(Z, leaves):
    K = number_of_objects(Z)
    if K == 0:
        #print "finished"
        return 1
    else:
        try:
            normalization_constant = 1./(1 - inclusion_prob(pull(Z,0), leaves))
        except:
            1/0
       
        prob = 0
        for ii in range(1,K+1):
           #print 'launching %dst of %d objects' %(ii, K)
           # print '%f' %inclusion_prob(pull(Z,ii),leaves)
            prob += inclusion_prob(pull(Z,ii),leaves)*object_map_prior(remove(Z,Z,ii),remove(leaves,Z,ii))
            
    return normalization_constant*prob
        
```

#### a function for making the sampling procedure more robust

```python
##shift the sample points
def circular_shift_by_n(samples,n,max_sample):
    return mod(samples+n,max_sample)
   
```

```python


##test with downsampling
pb = []
num = []
for ss in arange(0, 3000, 13):
    for stride in arange(2000, 25000, 500):
        samples = circular_shift_by_n(arange(0,250000,stride), ss, 250000)
        om = object_map[samples,:]
        rl = real_leaves[samples,:]
        rl_id = rl.any(axis=0)
        rl = rl[:, rl_id]
        om_id = om.any(axis=0)
        om = om[:,om_id]
        pb.append(object_map_prior(om, rl))
        num.append(om.shape[0])
plot(num, pb, '.')
xlabel('number of sampling points')
ylabel('object map probability')
```

##### this is how the probability of a single object map scales with number of sample points when there are < 10K leaves. as expected, probabilities are higher at lower resolution. note that at low res, massive dependence upon the location of the samples. so there is a need for robust sampling

#### try to estimate how long it will take to evaluate the probability of a reasonable number of objects

```python
import numpy as np
import time
import math

def nCr(n,r):
    f = math.factorial
    return f(n) / f(r) / f(n-r)

def stirling_num_of_2nd_kind(number_of_elements, number_of_sets):
    import math
    n = number_of_elements
    k = number_of_sets
    cnt = 0.
    for j in range(k+1):
        cnt += (-1)**(k-j)*nCr(k,j)*(j**n)
    return cnt/math.factorial(k)


K=5
P = 24
x = np.random.randint(1,2,size = (K**2,1))
y = np.random.randint(1,2,size = (K**2,200000*(K**2)))
tic = time.time()
np.sum(np.array(x==y).all(axis=0))
toc = time.time() - tic
print 'Elapsed: %.10f seconds' %(toc)
print 'Estimate # of minutes: %.10f' %(math.factorial(5)*toc/P*stirling_num_of_2nd_kind(9,K)/60)
```
