```python
import numpy as np
from PIL import Image
from matplotlib.pyplot import plot, imshow, show
%pylab inline
```

```python
sum_of_yeses = np.zeros((500,500))

for i in range(2000):
    im = Image.open('/musc.repo/Data/katie/testpics4_probes/cloud_pics_003042/cloud_imsize-500_cloud_num-100_' + str(i) + '.png') 
    mask = Image.open('/musc.repo/Data/katie/masks/003042.png')
    np_im = np.array(im)
    for i in range(500):
        for ii in range(500):
            if np_im[i,ii,0] != 0:
                np_im[i,ii,2] = 0
            if np_im[i,ii,1] != 0:
                np_im[i,ii,2] = 0
    im_blue = np_im[:,:,2]
    im_blue[np.where(im_blue < 254)] = 0
    indices_array = np.where(im_blue == 255)
    probes = zip(list(indices_array[1]), list(indices_array[0]))
    pix_vals = [mask.getpixel((int(x[0]), int(x[1]))) for x in probes]
    bool_val= len( set( pix_vals ) ) == 1
    if bool_val == True:
        sum_of_yeses += im_blue

        
imshow(sum_of_yeses)

    
    
```
