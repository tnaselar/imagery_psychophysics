#### create environment

```python
import numpy as np
from PIL import Image
from object_parsing.src.image_objects import view_mask
from imagery_psychophysics.src.mixture_of_histograms import mix_hog_image, plain_plot
from matplotlib.pyplot import plot, imshow, show
%pylab inline
```

#### read and parse datafile

```python
log_file = open('/musc.repo/Data/tnaselar/imagery_psychophysics/katie_2013_Nov_07_1508.log', 'r')
log_data = log_file.readlines()

probe_images_path = '/musc.repo/Data/tnaselar/imagery_psychophysics/cloud_pics/'

probe_image_size = (32,32)

V = np.prod(probe_image_size)

ovp = 29 #value of the "ON" pixels in the probe images -- had to find this by reading in the images


##get stimuli
stim_string = 'Set image_2 image=/home/katie/Desktop/testpictures/cloud_pics/'
probe_image_file = []
for sublist in log_data:
  if stim_string in sublist:
   probe_image_file += [sublist[sublist.find('image=')+50:-1]]
    

##get corresponding keypresses
keyp_string = 'Keypress:'
keypress = []
for sublist in log_data:
  if keyp_string in sublist:
    sub_string = sublist[sublist.find('Keypress')+10]
    if sub_string=='n' or sub_string=='y':
      keypress += sub_string

##get the "yeses"
yeses = [ii for ii,ss in enumerate(keypress) if ss=='y']
M = len(yeses)

##read-in all the image
probes = np.zeros((V,M))
for ii,yy in enumerate(yeses):
  foo = np.array(Image.open(probe_images_path+probe_image_file[yy]).convert('L').resize(probe_image_size))&ovp/ovp
  probes[:,ii] = foo.reshape(V)
  

##need to show the original!
    
##view the sum of all probes
imshow(probes.sum(axis=1).reshape(probe_image_size))

```

#### construct and train a model of the mental image

```python
##model parameters
K = 5 ##<<assumed number of objects in the scene
UNIFORMITY = 1
SMOOTHING = .000001

##initialize prior and likelihood distributions
naive_prior_dist = np.random.dirichlet(UNIFORMITY*np.ones(K)).reshape((1,K))
naive_lkhd_dist = np.random.dirichlet(np.ones(V), size=K).T

##create new model
mhi = mix_hog_image(naive_prior_dist, naive_lkhd_dist)

##EM
mhi.smoothing_param = SMOOTHING
mhi.run_em(probes)
```
