```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = {}
data['TN'] = {}
data['TN']['sourcefile'] = '/musc.repo/Data/katie/nbd_exp_data/TN_2014_Sep_23_1424.log'
data['TN']['perception'] = []
data['TN']['imagery'] = []
data['KL'] = {}
data['KL']['sourcefile'] = '/musc.repo/Data/katie/nbd_exp_data/katie_2014_Sep_24_1004.log'
data['KL']['perception'] = []
data['KL']['imagery'] = []


probe_images_path = '/musc.repo/Data/katie/nbd_probes/finalprobeset-%d/'
max_images_per_exp = 2

im_paths = []
for ii in range(1, max_images_per_exp+1):
    im_paths += [probe_images_path %(ii)]

imagery_marker = 'just-probes'
perception_marker = 'probes-with-im'


```

```python
for subject in data.keys():
    print subject
    for im in im_paths:
        print im
        stim_string = ' image=' + im
        log_data = open(data[subject]['sourcefile'], 'r').readlines()
        while log_data:
          sublist = log_data.pop(0)
          if stim_string in sublist:
            probe = sublist[sublist.find('image=')+55:-1]
            keypress_string = log_data.pop(1)
            if 'escape' in keypress_string:
                continue
            resp = int(keypress_string[-2])
            if imagery_marker in sublist:
                img_or_pcp = 'imagery'
            elif perception_marker in sublist:
                img_or_pcp = 'perception'
            data[subject][img_or_pcp] += [(probe, resp)]
```

```python
##sorted response
max_resp = 6+1
num_reps = 2
for ii,subj in enumerate(data.keys()):

    perception_consistency = np.zeros(max_resp)
    perception_resp_count = np.zeros(max_resp)

    imagery_consistency = np.zeros(max_resp)
    imagery_resp_count = np.zeros(max_resp)

    pcp = sorted(data[subj]['perception'], key=lambda tup: tup[0])
    nextel = pcp.pop(0)[1]
    while pcp:
        curel,nextel = nextel,pcp.pop(0)[1]
        perception_consistency[int(curel)] += curel==nextel
        perception_resp_count[int(curel)] += 1
    
    pcp = sorted(data[subj]['imagery'], key=lambda tup: tup[0])
    nextel = pcp.pop(0)[1]
    while pcp:
        curel,nextel = nextel,pcp.pop(0)[1]
        imagery_consistency[int(curel)] += curel==nextel
        imagery_resp_count[int(curel)] += 1
    plt.subplot(1,2,ii+1, title='%s' %(subj))
    plt.plot(range(0,max_resp), perception_consistency/perception_resp_count, 'b')
    plt.plot(range(0,max_resp), imagery_consistency/imagery_resp_count, 'r')
    plt.legend(('pcp', 'img'))
    plt.xlabel('response')
    plt.ylabel('consistency')
    plt.ylim((0,1))
    plt.xlim((0,max_resp))
    
    
```
