```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
subjects = {}
subjects['TN'] = {}
subjects['TN']['sourcefile'] = '/musc.repo/Data/katie/nbd_exp_data/TN_2014_Sep_23_1424.log'

subjects['KL'] = {}
subjects['KL']['sourcefile'] = '/musc.repo/Data/katie/nbd_exp_data/katie_2014_Sep_24_1004.log'


probe_images_path = '/musc.repo/Data/katie/nbd_probes/finalprobeset-%d/'
max_images_per_exp = 2

im_paths = []
for ii in range(1, max_images_per_exp+1):
    im_paths += [probe_images_path %(ii)]

imagery_marker = 'just-probes'
perception_marker = 'probes-with-im'

subject_table = pd.DataFrame(subjects)

number_of_lines_in_chunk = 3
```

```python
def starts_a_response_chunk(x):
    if 'New trial' in x:
        return True
    else:
        return False

def get_probe_number(new_line):   
    dx = new_line.index('.jpeg')
    if new_line[dx-2].isdigit():
        return int(new_line[dx-2:dx])
    else:
        return int(new_line[dx-1])
    
def get_repetition(new_line):
    return int(new_line[new_line.index('rep=')+4])

def get_state(new_line):
    if 'just-probes' in new_line:
        return 'img'
    elif 'probes-with-im' in new_line:
        return 'pcp'
    else:
        raise Exception('you are not at the start of a new trial')

def get_time_stamp(new_line):
    return float(new_line[0:new_line.index('\t')])

def get_response(new_line):
    try:
        return int(new_line[new_line.index('Keypress')+10])
    except:
        print('subject did not respond, returning None')
        return None
        
def skip_a_line(all_lines):
    return all_lines.pop(0)

def get_image(new_line):
    dx = new_line.index('finalprobeset')
    return new_line[dx:(dx+15)]
```

```python
print get_time_stamp('69.0318\tEXP\tStarted presenting image_2\n')
print get_state("68.9540\tEXP\tNew trial (rep=0, index=0): {'Stim_1': u'/musc.repo/Data/katie/nbd_probes/finalprobeset-1/just-probes/probe-65.jpeg'}\n")
print get_response('75.2320\tDATA\tKeypress: 4\n')
print get_image("68.9540\tEXP\tNew trial (rep=0, index=0): {'Stim_1': u'/musc.repo/Data/katie/nbd_probes/finalprobeset-1/just-probes/probe-65.jpeg'}\n")
```

```python
data_dict = {'subj': [], 'image': [], 'probe': [], 'rep': [], 'state': [], 'image_on': [], 'resp_on': [], 'response': []}
for subj in subjects.keys():
    print subj    
    all_lines = open(subjects[subj]['sourcefile'], 'r').readlines()
    while all_lines:
        new_line = all_lines.pop(0)
        if starts_a_response_chunk(new_line):
            data_dict['subj'].append(subj)
            data_dict['probe'].append(get_probe_number(new_line))
            data_dict['rep'].append(get_repetition(new_line))
            data_dict['state'].append(get_state(new_line))
            _ = skip_a_line(all_lines)
            data_dict['image'].append(get_image(all_lines.pop(0)))
            data_dict['image_on'].append(get_time_stamp(all_lines.pop(0)))
            new_line = all_lines.pop(0)
            data_dict['resp_on'].append(get_time_stamp(new_line))
            data_dict['response'].append(get_response(new_line))
            


probe_exp = pd.DataFrame(data_dict)

```

```python
def check_equal(x):
       return len(set(x)) <= 1
consistency = probe_multi.response.apply(check_equal)
first_probe = probe_multi.response.first()
```

```python


con = pd.concat([consistency, first_probe],axis=1)
con.columns = ['consistent', 'first_resp']
#con['first_response'] = first_probe
con.set_index('first_resp', append=True, inplace=True)
agg = con.groupby(level=['subj', 'state'])
fig = plt.figure()
plt.subplot(1,2,1)
agg.consistent.mean().plot(kind='bar')
cnt = 0
culuz = ['r','y', 'b', 'c']
plt.subplot(1,2,2)
for key,grp in agg:
    #print key
    #print grp
    #ax = plt.subplot(2,2,cnt, title=key)
    con_cur = grp.groupby(level=['first_resp']).consistent.apply(lambda x: 1/np.mean(x))
    ax = con_cur.plot(label=key, style = culuz[cnt])
    ax.set_ylabel('noise')
    
    cnt += 1
ax.legend(loc='upper right')    
    #plt.plot()
#for index, group in df.groupby(['model']):
#    group_agg = group.groupby(['training_examples']).aggregate(np.mean)
#    group_agg.plot(y='accuracy', label=index)
```
