### establish enivornment

```python
##establish multi-engine client
from IPython.parallel import Client
dview = Client()[:]
dview.use_dill()
%load_ext autoreload
%autoreload 2
```

```python
%%px --local
import numpy as np
from imagery_psychophysics.src.model_z import model_z
from imagery_psychophysics.src.model_z import noise_grid
from imagery_psychophysics.src.probes import probes
from imagery_psychophysics.src.stirling_maps import sparse_point_maps
from imagery_psychophysics.src.object_map import object_map as om
from imagery_psychophysics.src.poisson_binomial import poisson_beta_binomial as pbb
from PIL.Image import fromarray
import time
```

```python
##local only
%matplotlib inline
import matplotlib.pyplot as plt
```

### distribute key parameters

```python
%%px --local
K = 4 ##the number of clusters in what will be the target map
on_off_prob = (.87,.05)
seed_point_rows = 3
seed_point_cols = 3
size_of_field = 32
```

### create probes

```python
%%px --local
probe_set = probes(size_of_field=size_of_field)
probe_set.masks = probe_set.masks.reshape((probe_set.num_probes, size_of_field, size_of_field))
```

### generate a simulated target map

```python
%%px --local
spm = sparse_point_maps(seed_point_rows, seed_point_cols, size_of_field, cluster_pref='all', number_of_clusters = K)
spm.scatter()
target_map = spm.nn_interpolation(dx=[1000])
target_image = fromarray(np.squeeze(target_map).astype('uint8'))
target_om = om(target_image)
```

```python
##see it
num_samples = target_om.count_myself()*20
good_samples = target_om.good_samples(num_samples)
plt.imshow(target_om.map_as_array)
plt.scatter(good_samples[:,0], good_samples[:,1],c='k')
```

### generate noisy responses to target map

```python
%%px --local
lkhd = pbb(K, on_off_prob = on_off_prob)
target_noisy_resp = lkhd.sample(target_om.overlap(probe_set.masks))
```

```python
##see the noisy responses
resp_shape = (np.sqrt(probe_set.num_probes), np.sqrt(probe_set.num_probes))
plt.imshow(target_noisy_resp.reshape(resp_shape).T)
plt.colorbar()
```

```python
##and the likelihood is:
print lkhd.sum_log_likelihood(target_noisy_resp, target_om.overlap(probe_set.masks))
```

### instantiate a model_z (mental_image)

```python
%%px --local
init_object_map = np.random.randint(1,K+1, size = (size_of_field, size_of_field))
start = time.time()
mental_image = model_z(K, probe_set.masks, target_noisy_resp, size_of_field=size_of_field, on_off_prob = on_off_prob, om = init_object_map)
print 'seconds: %f' %(time.time()-start)
```

### co-optimize object map and noise parameters

```python


##define a function for optimizing on each loop: everything is defined locally, so no need for arguments?
def optimize_likelihood(cur_noise):
    print 'best noise: %f,%f' %(cur_noise)
    candidate_lkhd = map(lambda x: mental_image.sum_log_lkhd(x, on_off_prob=cur_noise), maps)
    cur_idx = np.argmax(candidate_lkhd)
    print 'best score: %f' %(candidate_lkhd[cur_idx])
    cur_map = np.squeeze(maps[cur_idx,:,:])
    cur_dict = dict(cur_map = cur_map)
    dview.push(cur_dict)
    print 'best index: %d' %(cur_idx)
    candidate_lkhd = dview.map(lambda x: mental_image.sum_log_lkhd(cur_map,x), zip(on_prob,off_prob))   
    cur_idx = np.argmax(candidate_lkhd)
    print 'best score: %f' %(candidate_lkhd[cur_idx])
    cur_noise = (on_prob[cur_idx], off_prob[cur_idx])
    cur_dict['cur_noise'] = cur_noise
    dview.push(cur_dict)
    return candidate_lkhd[cur_idx], cur_map, cur_noise

```

```python
#create noise grid
dview.block = True
noise_grid_x_dns,noise_grid_y_dns = 20,20
on_prob, off_prob = noise_grid(noise_grid_x_dns,noise_grid_y_dns)
#initialize scores and noise_params
current_score, previous_score = -10**7,-float('inf') ##two smallest numbers
cur_noise = (mental_image.noise_model.on_prob, mental_image.noise_model.off_prob)
cur_dict = dict(cur_noise=cur_noise)
dview.push(cur_dict)
maps = spm.nn_interpolation()
while current_score > previous_score:
    previous_score = current_score
    start = time.time()
    current_score, current_map, cur_noise = optimize_likelihood(cur_noise)
    print 'seconds: %f' %(time.time()-start)

##reinstantiate with optimal parameters. only local needs to know about the update
mental_image = model_z(K, probe_set.masks, target_noisy_resp, size_of_field=size_of_field, on_off_prob = cur_noise, om = current_map)
```

```python
plt.subplot(1,2,1, title = 'best guess')
plt.imshow(mental_image.best_om)
plt.axis('off')
plt.subplot(1,2,2, title = 'target')
plt.imshow(target_image)
plt.axis('off')
```

```python
##test prediction performance
from scipy.stats import pearsonr 
probe_set_shape = ([np.sqrt(probe_set.masks.shape[0])]*2)

##predicted response to best stirling map chosen during model init.
pred_resp = mental_image.predict_response(probe_set.masks)
r,p = pearsonr(mental_image.resp, pred_resp)
print r
print p
plt.subplot(1,3,1, title='predicted responses')
plt.imshow(pred_resp.reshape((probe_set_shape)))
plt.axis('off')

##optimal predicted responses from the target map used to produce the fake noisy data
ideal_resp = pbb(K, on_off_prob = (1,0)).sample(target_om.overlap(probe_set.masks))
r,p = pearsonr(mental_image.resp, ideal_resp)
print r
print p
plt.subplot(1,3,2, title='ideal responses')
plt.imshow(ideal_resp.reshape((probe_set_shape)))
plt.axis('off')

##actual fake responses
plt.subplot(1,3,3, title='actual responses')
plt.imshow(mental_image.resp.reshape((probe_set_shape)))
plt.axis('off')
```
