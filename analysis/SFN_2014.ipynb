#### make pictures of the response maps

```python
##make pictures of response maps
exp_agg = response_data.groupby(by=['subj', 'image','state'])
```

```python
def response_map(name,grp):
    mx = max(grp.probe.apply(lambda x: int(x.split('_')[0])))
    pp = np.zeros((known_probe_size,known_probe_size))
    image = name[1]
    print name
    for ii in range(mx+1):
        probe_1 = str(ii)
        r1 = int(grp[grp['probe']== probe_1].response)
        dx  = probes[image]['index'].index([ii])
        pp += np.array(probes[image]['mask'][dx]>0, dtype='int')*r1
    return pp
```

```python
##view: construct an image grid
fig = plt.figure(1, (30,30))
grid = ImageGrid(fig, 111, # similar to subplot(111)
                nrows_ncols = (6, 2), # creates 2x2 grid of axes
                axes_pad=0.5, # pad between axes in inch.
                cbar_mode = 'each',
                cbar_pad = .05
                )

##order the response maps for eacse of comparison
cnt = 0
K = response_data.response.max()-1
cmap2 = plt.cm.get_cmap('Blues', K) 
for name,grp in exp_agg:
    img = '/musc.repo/Data/tnaselar/imagery_psychophysics/poly_probes/target_images/'+name[1]+'_image.png'
    rmap = response_map(name,grp)
    grid[cnt].imshow(Image.open(img))
    im = grid[cnt].imshow(rmap, interpolation='nearest', alpha = .7, cmap=cmap2, clim=[1,K+1])
    grid[cnt].set_title(name)
    grid[cnt].set_xticks([])
    grid[cnt].set_yticks([])
    #grid.cbar_axes[cnt].colorbar(im,ca)
    
    grid[cnt].cax.colorbar(im)
    grid[cnt].cax.set_yticks(range(1,K+1,1))
    #plt.colorbar(im, ax=grid[cnt])
    #plt.axis('off')
    cnt += 1
    plt.savefig('/musc.repo/Data/tnaselar/imagery_psychophysics/poly_probes/results/response_maps.png', transparent = True)
```

#### analysis of noise parameters

```python
def noise_params(name,grp):
    noise_agg = grp.groupby(['capacity', 'resolution'])
    cnt = 0.
    noise_marge = np.zeros((noise_agg.p_on.size().max(),1))
    for noise_name,noise_grp in noise_agg:
        noise_marge += noise_grp[['marginal']]
        cnt += 1
    p_on = noise_grp.p_on
    p_off = noise_grp.p_off ##<<just grab from the last one, they're all the same
    return noise_marge/cnt, p_on, p_off


        
```

```python
marge_agg = marginals.groupby(['subj', 'image', 'state'])
fig = plt.figure(1,figsize=(20,20))
cnt = 1
cmap2 = plt.cm.get_cmap('gist_heat', 30)
for name,grp in marge_agg:
    mat,pon,poff = noise_params(name,grp)
    ax = fig.add_subplot(6,2,cnt, aspect=1.0, axisbg=np.array([84, 149, 178] )/400.)#13,76,85])/350., ) #'gray')##[13,76,85]
    mappable=ax.scatter(pon,poff,s=1500,c=np.log(mat), cmap=cmap2, marker='o',linewidth=0,alpha=.9)
#     plt.axis('equal')
    plt.xlim([0,1.2])
    plt.ylim([0,1.2])
    plt.xticks([])
    plt.yticks([])
#     plt.xlabel('p_on')
#     plt.ylabel('p_off')
#     plt.axis('tight')
    plt.colorbar(mappable)
    plt.title(name)
    cnt += 1
# fig.subplots_adjust()    
fig.tight_layout()
plt.savefig('/musc.repo/Data/tnaselar/imagery_psychophysics/poly_probes/results/noise_params.png')
```

```python
def capacity_param(name,grp):
    noise_agg = grp.groupby(['p_on', 'p_off', 'resolution'])
    cnt = 0.
    noise_marge = np.zeros((noise_agg.capacity.size().max(),1))
    for noise_name,noise_grp in noise_agg:
        noise_marge += noise_grp[['marginal']]
        cnt += 1
    capacity = noise_grp.capacity
    return noise_marge/cnt, capacity
```

```python
def resolution_param(name,grp):
    noise_agg = grp.groupby(['p_on', 'p_off', 'capacity'])
    cnt = 0.
    noise_marge = np.zeros((noise_agg.resolution.size().max(),1))
    for noise_name,noise_grp in noise_agg:
        noise_marge += noise_grp[['marginal']]
        cnt += 1
    resolution = noise_grp.resolution
    return noise_marge/cnt, resolution
```
