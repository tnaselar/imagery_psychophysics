```python
import numpy as np
from scipy.misc import comb as nCk
from matplotlib import pyplot as plt
from imagery_psychophysics.src.model_z import noise_grid
from operator import mul as mul
from imagery_psychophysics.src.stirling_maps import stirling_num_of_2nd_kind as snk
import pandas as pd
from scipy import optimize as opt
%matplotlib inline
```

```python
def counts(r,d,n):
    return np.array([nCk(d,m)*nCk(n-d, r-m) for m in range(min(r,d)+1)])
```

```python
def lkhd(r,d,n, prob_on, prob_off):
    ratio_on = np.array([((prob_on)/(1-prob_on))**m for m in range(n+1)])
    #print ratio_on
    ratio_off = np.array([((prob_off)/(1-prob_off))**(r-m) for m in range(n+1)])
    #print ratio_off
    return (1-prob_on)**d * (1-prob_off)**(n-d) * counts(r).dot(ratio_on*ratio_off)
    
```

```python
def lkhd(r,d,n,p_on,p_off):
    probs = np.array([(1-p_on)**(d-m) * (p_on)**m * (p_off)**(r-m) * (1-p_off)**(n-d-r+m) for m in range(min(r,d)+1)])
    #print probs
    return counts(r,d,n).dot(probs)
```

```python
##this is a very flexible distribution

n = 5
d = 3
p_on, p_off = noise_grid(5,5)
for p in zip(p_on,p_off):
    probs = [lkhd(r,d,n, p[0],p[1]) for r in range(n+1)]
    ##sanity check
    if np.abs(np.sum(probs)-1) > .00000000000001:
        print 'uh-oh'
        print np.sum(probs)
    plt.plot(range(n+1), probs,label = '(%0.2f, %0.2f)' %(p))

#plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=4, fancybox=True, prop = {'size': 8})

plt.xlabel('r')
plt.ylabel('p(r | d,n,p_on,p_off)')
plt.title('d = %d' %(d))
```

```python
##when the noise parameters are reasonable (i.e., prob_off < 0.5 < prob_on), max posterior is at r=d
##when noise params are pathological: prob_off > 0.5, or prob_on < 0.5 , max posterior heads toward either n or 0, resp.
n = 5
d = 3
p = (0.8, 0.17)
plt.plot(range(n+1), [lkhd(r,d,n, p[0],p[1]) for r in range(n+1)], label = '(%0.2f, %0.2f)' %(p))
p = (.6, 0.45)
plt.plot(range(n+1), [lkhd(r,d,n, p[0],p[1]) for r in range(n+1)], label = '(%0.2f, %0.2f)' %(p))
p = (0.53, 0.49)
plt.plot(range(n+1), [lkhd(r,d,n, p[0],p[1]) for r in range(n+1)], label = '(%0.2f, %0.2f)' %(p))

##pathological: p_on is less than 0.5
p = (0.44, 0.13)
plt.plot(range(n+1), [lkhd(r,d,n, p[0],p[1]) for r in range(n+1)], label = '(%0.2f, %0.2f)' %(p))

##pathological: p_off is greater than 0.5
p = (0.88, 0.74)
plt.plot(range(n+1), [lkhd(r,d,n, p[0],p[1]) for r in range(n+1)], label = '(%0.2f, %0.2f)' %(p))
plt.legend(loc = 'upper left', ncol = 2, prop = {'size': 8})
plt.xlabel('r')
plt.ylabel('posterior (over r)')
```

```python
##as a consequence, max lkhd depends upon whether r = d

n = 5
d = 3

##assuming p_off reasonable,
##for r < d ,  p_on tends toward p_off. for r >= d, p_on = 1.
p_off = 0.1
p_on = np.linspace(p_off+.001, 0.999, num=25)
for r in range(n+1):
    plt.plot(p_on, [lkhd(r,d,n,p,p_off) for p in p_on], label = 'r = %d' %(r))
plt.ylabel('likelihood')
plt.xlabel('p_on')
plt.legend(loc = 'upper center', ncol = 2, prop = {'size': 8})
plt.title('p_off = %0.2f, d = %d' %(p_off,d))  
```

```python
##similarly, assuming p_on reasonable, max lkhd value of p_off tends toward 0 as r --> d; tends toward 1 as r--> n
p_on = 0.999
p_off = np.linspace(0+.001, p_on-.0001, num=25)
plt.subplot(2, 1, 2)
for r in range(n+1):
    plt.plot(p_off, [lkhd(r,d,n,p_on,p) for p in p_off], label = 'r = %d' %(r))
plt.ylabel('likelihood')
plt.xlabel('p_off')
plt.legend(loc = 'upper center', ncol = 3, prop = {'size': 8})
plt.title('p_on = %0.2f, d = %d' %(p_on,d)) 
```

```python
##a reasonable multivariate version takes about 70ms for every 1000 datapoints (on ichi)
##so evaluations for the e-step will be in 10 thousands ~ 1 sec.
##evaluations for the m-step will be in the 1000's ~ 
def multi_lkhd(r,d,n,p_on,p_off):
    #return np.log(reduce(mul, np.array(map(lambda x,y: lkhd(x,y,n,p_on,p_off), r,d))))
    return np.sum(np.array(map(lambda x,y: np.log(lkhd(x,y,n,p_on,p_off)), r,d)))

n, p_on, p_off = 5, .7, .2
T = 7000
r = np.random.randint(1, n+1, T)
d = np.random.randint(1, n+1, T)
%timeit multi_lkhd(r,d,n, p_on,p_off)
print multi_lkhd(r,d,n, p_on,p_off)
```
