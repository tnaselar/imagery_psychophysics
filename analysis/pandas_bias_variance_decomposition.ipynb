```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from random import choice
from itertools import tee, izip
%matplotlib inline
```

```python
probe_exp = pd.load('/musc.repo/scratch/data.pkl')
```

```python

##group the data so we can analyze responses to each probe individually at first
probe_multi = probe_exp.groupby(['subj', 'image', 'state', 'probe'])
```

```python
##an example of how to get data from the "groupby" object
keys = probe_multi.groups.keys()
dx = probe_multi.groups[keys[0]]
probe_exp.iloc[dx]
```

```python
##to measure consistency we want to see if the response across two reps are equal
def check_equal(x):
       return len(set(x)) <= 1

##get consistency for responses in each group (defined by the "groupby" statement above)    
consistency = probe_multi.response.apply(check_equal)

##we also want to know what was the response to the first probe
first_probe = probe_multi.response.first()
```

```python
##put the consistency measure together with the first probe response
##the next 2 lines give us a dataframe with "consistent" and "first_resp" as columns
##this dataframe will have hierarchical index (a MultiIndex)
con = pd.concat([consistency, first_probe],axis=1)
con.columns = ['consistent', 'first_resp']

##pivot the "first_resp" column into the index. now the MultiIndex has another level.
con.set_index('first_resp', append=True, inplace=True)

##we now regroup, this time using levels of the MultiIndex to define the groups
##in this line, we group together all of the consistency measures belonging to one subject/state
agg = con.groupby(level=['subj', 'state'])

##create a plot of the mean consistency for each subject for imagery/perception
fig = plt.figure()
plt.subplot(1,2,1)
ax = agg.consistent.mean().plot(kind='bar')
ax.set_ylabel('consistency')

##now we will look at consistency across values of the first response to see if there 
##is any obvious relationship between variance and mean response
cnt = 0
culuz = plt.cm.BrBG(np.array([.3,.7,.1,.9]))
plt.subplot(1,2,2)
for key,grp in agg:
    #print key ##<<this in the index for each of the iterated groups
    #print grp ##<<this a Series (or maybe a DataFrame) corresponding to each group in agg
    ##group again, this time grouping all consistency measures according to first response, and take an average
    con_cur = grp.groupby(level=['first_resp']).consistent.apply(np.mean)
    ax = con_cur.plot(label=key, color = culuz[cnt,:])
    ax.set_ylabel('consistency')
    cnt += 1
ax.legend(loc='lower right', prop={'size': 8})    
plt.tight_layout(pad = 4)
plt.savefig('consistency.png')
```

```python
##bias variance decomposition

##group by subject, image, state
bv = probe_exp.groupby(['subj', 'image'])
```

```python
##now split into subj/image pairs of pcp and img states
sample_size = 10000
bv_results = dict()
for key,grp in bv:
    #print key
    group_by_state = grp.groupby(['state'])
    try:
        pcp_dx = group_by_state.groups['pcp']
        img_dx = group_by_state.groups['img']
    except:
        print 'missing data'
        continue
    pcp_resamples = probe_exp.iloc[pcp_dx].groupby(['probe']).response.apply(lambda x: np.random.choice(x,size = sample_size))
    img_resamples = probe_exp.iloc[img_dx].groupby(['probe']).response.apply(lambda x: np.random.choice(x,size = sample_size))
    pcp_mean = pcp_resamples.apply(np.mean)
    img_mean = img_resamples.apply(np.mean)
    
    pcp_error = np.mean((pcp_resamples - pcp_mean).apply(lambda x: np.mean(x**2)))
    #print pcp_mean_error
    pcp_var = pcp_error
    pcp_bias = 0
    
    img_error = np.mean((img_resamples - pcp_mean).apply(lambda x: np.mean(x**2)))
    img_var = np.mean((img_resamples - img_mean).apply(lambda x: np.mean(x**2)))
    #print img_var
    img_bias = np.mean((img_mean-pcp_mean)**2)
    #print img_error
    #print img_bias + img_var
    pcp_key = list(key)+['pcp']
    print pcp_key
    img_key = list(key)+['img']
    bv_results[tuple(pcp_key)] = [pcp_var, pcp_bias]
    bv_results[tuple(img_key)] = [img_var, img_bias]
    
    
    #print np.array(pcp_mean).shape
    #print img_bias+img_var
    #print img_error
#keys = bv.groups.keys()
#dx = bv.groups[keys[1]]
```

```python
foo = pd.DataFrame(bv_results, index = ['var','bias']).transpose()
#foo  = foo.groupby(level = 2)
ax = foo.plot(kind = 'bar', stacked = 'True', colormap = 'Dark2', fontsize = 18)
ax.set_yticklabels('')
ax.set_ylabel('error (a.u.)')
ax.set_xticklabels(['KL\nimg1', 'KL\npcp1', 'KL\nimg2', 'KL\npcp2', 'TN\nimg1', 'TN\nimg2'], rotation='horizontal')
ax.legend(fontsize=12)
plt.savefig('bias_variance.png')
```
